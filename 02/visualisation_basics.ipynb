# Basics of visualization
One picture (or graph) can sometimes say more than a thousand words. This is even more the case for exploratory data analysis.

In this lesson, we'll show you how to plot some basic types of graphs: bar plot, line plot, and scatter plot.

## The diverse world of visualization libraries in Python
While there is a consensus on a library for common tabular data processing, and we usually look for `pandas` when examining small to medium-sized data of a non-exotic type, there are plenty of libraries for data visualization. Each has its advantages and disadvantages. We will introduce the following three in the course and we will focus mainly on how to use them together with pandas:
- `matplotlib` - This is probably the most widespread and in many ways the most flexible library. It is the default option if you need good-looking static graphs that will work almost anywhere. Considerable flexibility is balanced by sometimes not entirely intuitive names of functions and arguments. Pandas uses it internally (that&#39;s why you don&#39;t have to get acquainted with it in such detail). See https://matplotlib.org/.
- `seaborn` - The aim of this library is to help with statistical graphs in particular. It builds on matplotlib. We will work with it to visualize more complex relationships between multiple variables. See https://seaborn.pydata.org/.
- `plotly` (especially its subset` plotly.express`) - You will reach for this library if you want to add interactivity to your visualization. Especially when working in a Jupyter laptop, it allows you to explore everything much faster. See https://plot.ly/python/.
ðŸ’¡ For a more detailed explanation, we recommend watching the (slightly older) video by Jake Vanderplase: Python Visualizations&#39; Landscape (https://www.youtube.com/watch?v=FytuB8nFHPQ), which summarizes the basic features of each library.

Key things to remember about data visualisation in matplotlib:

* Matplotlib graphs your data on `Figures` (e.g. Jupyter Lab widgets)
* Each `Figure` can contain one or more `Axes`, an area where points can be specified in terms of x-y coordinates`

The simplest way of creating a`Figure` with an `Axes` is using pyplot.subplots. We can then use Axes.plot to draw some data on the Axes`

```python
width = 0.35
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(["Jan","Feb","Mar"], [500, 900, 4200], color ="blue" , label = "spending person B")
ax.plot(["Jan","Feb","Mar"], [590, 950, 200], color ="yellow" , label = "spending person B")
ax.bar(["Jan","Feb","Mar"], [900, 1300, 3400], color =["red", "red", "green"] , label = "spending person A")
ax.set_xlabel("here are xes", fontsize=20)
ax.legend()
# Plot some data on the axes.
```

### Two ways of plotting in Matplotlib

As noted above, there are essentially **two ways** to use Matplotlib:

1. Explicitly create `Figures` and `Axes`, and call methods on them (the "object-oriented (OO) style").
2. Rely on`pyplot` to automatically create and manage the`Figures` and `Axes`, and use `pyplot` functions for plotting.

### Object-oriented plotting

```python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend();  # Add a legend.
```

This is a hybrid object that can be called as a method (`Series.plot ()` - uses the default chart type), or you can use another dot to reference its own methods that draw different chart types. For pedagogical reasons, we want to start with a bar graph that is not the default, so we call [`Series.plot.bar ()`] (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas). Graphing functions offer a lot of parameters that are not well documented and are quite closely related to how the `matplotlib` library works. We will show them gradually when they come in handy. We would like to enlarge our chart a bit in height. Also, the values do not differ much from each other, and setting a custom range on the x-axis would help highlight the differences. Plus we add a little formatting.
- `figsize` specifies the size of the graph as a tuple of size in inches (width, height). To experiment with the ideal value, simply experiment on your laptop.- `xlim` specifies the range of values on the x-axis in a similar pair (minimum, maximum)- `color` specifies the fill color: it can be a name or a hexadecimal RGB notation- `edgecolor` says the color of the columns- `title` sets the title of the whole graph

```python
eu_countries["life_expectancy"].sort_values().plot.barh(
    figsize=(6, 8),
xlim = (75, 85),    color="yellow",
edgecolor = &quot;# 888888&quot;, # medium graytitle = &quot;Life expectancy (years)&quot;);
```

ðŸ’¡ Starting bar graphs (but also many others) other than at zero will help you notice even slight differences. This is definitely a good idea in the exploratory phase. However, when presenting the results, the highlighted differences can confuse the audience and give the impression that an effect is significantly stronger than it actually is. The less intuitive the presented data, the stronger the manipulation effect. In this case, probably few would believe that people live in Spain sixty times longer than in Latvia, because it does not meet normal expectations, but even so, at first glance the situation looks very dramatic (we leave you to judge whether the difference between 75 and 83, or about 10% is huge or not). This is how journalists confuse quite often - either intentionally or by mistake.

We can easily display multiple quantities in the graph at once, if we do not create it through `Series`, but` DataFrame`. All you have to do is supply more columns instead of one column (for example, by selecting from `DataFrame`) and we will see more columns below each row.
In our case, we will look at how many years men and women live. We choose gender stereotypical colors (it is sometimes more intuitive), but of course you can adjust them as you wish.

```python
eu_countries.sort_values("life_expectancy")[["life_expectancy_male", "life_expectancy_female"]].plot.barh(
    figsize=(8, 10),
xlim = (68, 88), # axis rangecolor = [&quot;blue&quot;, &quot;red&quot;], # two different colors for two columnsedgecolor = &quot;# 888888&quot;, # medium graytitle = &quot;Life expectancy (years)&quot;);
```

## Scatter plot
A scatter plot is the easiest way to compare two different quantities. In the coordinate system, as used in mathematics, each line corresponds to one point. The values of the two columns then encode the coordinates `x` and` y`. This is also reflected in the way we create the scatter plot in `pandas`.
We call the method [`plot.scatter`] (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html) of our table and supply it as arguments` x` and `y` column names to use for coordinates:

```python
# Relationship between drinking and life expectancycountries.plot.scatter(
    x="alcohol_adults",
    y="life_expectancy",
);
```

ðŸ’¡ We will talk about causality, correlations and connections between quantities at another time, but you also can&#39;t help the impression that the more you drink somewhere, the longer you live there?
Even without mathematical rigor, we know where the dog will be buried. Let&#39;s try to color individual regions of the world with different (stereotypical?) Colors. We will learn a handy method [`map`] (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.map.html), which will replace the values in` Series` according to the dictionary from-&gt; to (and returns a new `Series` instance). The `world_4region` column contains exactly 4 different areas (&quot; continents &quot;), so a very simple dictionary will suffice.
Here are some more arguments:- `s` expresses the size (or approximately the area) of the symbol in points (it can be one value or a column / field of values)- `marker` indicates the shape of the symbol, usually in one letter, see [list of options] (https://matplotlib.org/3.1.1/api/markers_api.html)- `alpha` expresses the opacity of the symbol (0 = completely transparent and not visible, 1 = opaque, intense, hides everything&quot; behind &quot;). It is useful when we have a large number of symbols in the chart and we want to allow them to overlap.

```python
# Relationship between drinking and life expectancycountries.plot.scatter(
    figsize=(7, 7),
    x="alcohol_adults",
    y="life_expectancy",
marker = &quot;h&quot;, # Symbol shape: hexagon - (h) exagons = countries [&quot;population&quot;] / 1e6, # Symbol size (per second) by populationedgecolor = &quot;black&quot;, # Border coloralpha = 0.5 # Semi-transparent symbols);
```

Some parameters may not have a common value for all rows. You can use a list or `Series` of the correct length instead, which we will now show in the color we adapt to the region (using the [` map`] method (https://pandas.pydata.org/pandas-docs/stable/reference/ api / pandas.Series.map.html)):

```python
region_colors = {    "europe": "blue",
    "asia": "yellow",
    "africa": "black",
    "americas": "red"
}
color = countries [&quot;world_4region&quot;]. map (region_cores)color
```

Now that we use `color` in the` color` argument, a specific value is applied to each point.

```python
countries.plot.scatter(
    figsize=(7, 7),
    x="alcohol_adults",
    y="life_expectancy",
marker = &quot;h&quot;, # Symbol shape: hexagon - (h) exagoncolor = color, # Unfortunately we can&#39;t just use the column name, we have to give the whole &quot;array&quot; of valuess = countries [&quot;population&quot;] / 1e6, # Symbol size (per second) by populationedgecolor = &quot;black&quot;, # Border coloralpha = 0.5 # Semi-transparent symbols);
```

And so it actually seems that people generally drink little, in America so moderately, in Africa people live to a younger age, but at first glance we do not see any trend in these groups of countries. The only continent that stands out is Europe, where people drink a lot and live a long time, but both are probably the result of the modern way of life. On closer inspection, on the contrary, it seems that within Europe, more drinking means a shorter life.

It often happens that the values are difficult to compare. For example, in terms of area or population, there are miniature and gigantic countries in the world. The differences are of the order:

## Line plot
This type of graph makes sense especially if one variable evolves continuously depending on another variable. Time series are a great example of this (whether for the relationship between time and a quantity, or two quantities that both evolve at the same time).
You create a line graph using the [`plot.line`] method (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.line.html). Coincidentally, this is also the default graph type for `pandas`, so it is enough to call` plot` as a method. It has parameters similar to `scatter`.
For example, let&#39;s look at the development of life expectancy in the Czech Republic, as it has developed over time since the early 1980s:

```python
czech.plot.line(x="year", y="life_expectancy");
```

Of course, we can draw more columns again.

```python
czech.plot(x="year", y=["life_expectancy_female", "life_expectancy_male"]);
```

There are several interesting arguments for line graphs:
- `lw` indicates the line thickness in points- `style` is a line style:&quot; - &quot;is solid,&quot;: &quot;dotted,&quot; - &quot;dashed,&quot; -. &quot; dashed- `markersize` is the size of the symbol that can optionally accompany the line

```python
czech.plot.line(
    x="year",
y = [&quot;bmi_men&quot;, &quot;bmi_women&quot;],lw=1,    style="--",
marker = &quot;o&quot;, # Add round points for values from the tablemarkersize = 3);
```

### Bonus 1: &quot;clean&quot; matplotlib
Because the default graphing in `pandas` builds on the` matplotlib` library and only the individual functions wrap and make it easier to work with columns, the function parameters will usually be similar (the main difference is that they do not take column names, you have to pass the column as such).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(7,7))

ax.scatter(
    countries["alcohol_adults"],
    countries["life_expectancy"],
    s=countries["population"] / 1e6,
    color=countries["world_4region"].map({"europe": "blue", "asia": "yellow", "africa": "black", "americas": "red"}),
    edgecolor="black",
marker = &quot;h&quot;,    alpha=0.5
);

# Axis labels must be filled in manuallyax.set_xlabel("alcohol_adults")
ax.set_ylabel("life_expectancy");
```
